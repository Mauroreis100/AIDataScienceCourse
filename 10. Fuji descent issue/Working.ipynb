```python
import numpy as np
csv_path = "mtfuji_data.csv" # Specify the file name (path)
np.set_printoptions(suppress=True) # Set prohibition of scientific notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
fuji[130:140]
```

```python
# Extract x (point index) and y (elevation)
elevation_data=fuji
x_points = elevation_data[:, 0]
y_elevation = elevation_data[:, 3]
print(y_elevation)
```

```python
import math
import matplotlib.pyplot as plt
def plot_elevation(x, y):

    plt.figure(figsize=(10,5))
    plt.plot(x,y, color="blue")
    plt.title("Mt. Fuji")
    plt.xlabel("position")
    plt.ylabel("elevation[m]")
    plt.show()

plot_elevation(x_points, y_elevation)
```

```python
def calculate_gradient(point, elevation_data):
    """
    Computes the gradient at a given point.
    Uses the previous point to estimate slope.
    """
    idx = np.where(elevation_data[:, 0] == point)[0]

    if len(idx) == 0 or idx[0] == 0:
        return None  # No gradient can be computed

    current_idx = idx[0]
    prev_idx = current_idx - 1

    x1, y1 = elevation_data[current_idx]
    x2, y2 = elevation_data[prev_idx]

    gradient = (y1 - y2) / (x1 - x2)  # Slope calculation
    return gradient

# Example usage
print(calculate_gradient(136, elevation_data))  # Example: Compute gradient at point 136
```
