# [Question 6] Creating a diagram
It is important to visualize and check the data. Please do the following three things. In addition to Matplotlib, seaborn is also commonly used as a plotting library.

* Create a pie chart showing the number of samples per label (including percentages)
* Select features one by one and visualize the distribution of data by label using box plots
* Select features one by one and visualize the distribution of data by label using a violin plot

Box plots and violin plots display similar figures. Explain how they differ and what advantages each has.

```python
#Create a pie chart showing the number of samples per label (including percentages)
# Import libraries
from matplotlib import pyplot as plt
import numpy as np

feature = ['sepal_length', 'sepal_width', 'petal_length',
        'peta_width', 'Species']

sample_number = [df['sepal_length'].count().sum(), df['sepal_width'].count().sum(), df['petal_length'].count().sum(),df['peta_width'].count().sum(), df['Species'].count().sum()]


# Creating plot
fig = plt.figure(figsize=(3, 5))
plt.pie(sample_number, labels=feature, autopct='%1.1f%%')

# show plot
plt.show()
```

```python
#Select features one by one and visualize the distribution of data by label using box plots
import seaborn as sns
# Box plot for a single feature (e.g., sepal length)
plt.figure(figsize=(8, 5))
sns.boxplot(x='label', y='sepal length (cm)', data=data, palette="Set2")
plt.title("Box Plot: Sepal Length by Label")
plt.show()
```
