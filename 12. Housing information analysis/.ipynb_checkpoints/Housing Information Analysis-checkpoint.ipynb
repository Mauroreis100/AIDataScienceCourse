Check what percentage of data is missing.

```python
# a simple visualization of nullity by column:
import missingno as msno
%matplotlib inline
msno.bar(data)
```

```python
msno.heatmap(data)
```

```python
data.isnull().sum()
```

```python
msno.matrix(data.sample(500))
```

```python
total = data.isnull().sum().sort_values(ascending=False) # Total rows that are null in each collumn
percent = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False) # Percentage of what's missing and total
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'missing_ratio'])
missing_data
```

Features (columns) with five or more missing values ​​are deleted.

```python
# Delete features (columns) that have 5 or more missing values.
pct_null = data.isnull().sum()
missing_features = pct_null[pct_null >= 5].index
data.drop(missing_features, axis=1, inplace=True)

# Samples (rows) with missing values are deleted from the data from which features with 5 or more missing values have been deleted.
data.dropna(axis=0, thresh=5, inplace=True)
```

From the data after removing features with five or more missing values, samples (rows) with missing values ​​are deleted.

# [Question 5] Terminology research
You need to know the following two words. Research and summarize them.

* kurtosis
* skewness

1. Kustoris - indicates how the tails of a distribution differ from the normal distribution. Use kurtosis to help you initially understand general characteristics about the distribution of your data. High kurtosis implies more extreme values in the distribution, while low kurtosis indicates a flatter distribution.

2. Skewness - is the extent to which the data are not symmetrical. Whether the skewness value is 0, positive, or negative reveals information about the shape of the data and about its mean.

# [Problem 6] Checking the distribution


We use seaborn sns.displot()'s and sns.histplot()to display the distribution of the dependent variable, 

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.displot(data['SalePrice']);
```

```python
sns.histplot(data['SalePrice'])
```

and also calculate "kurtosis" and "skewness".

```python
#skewness and kurtosis
print("Skewness: %f" % data['SalePrice'].skew())
print("Kurtosis: %f" % data['SalePrice'].kurt())
```

```python
# Exclude non-numeric columns before calculating correlation
numeric_df_train = data.select_dtypes(include=['int64', 'float64'])
corrmat = numeric_df_train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
import numpy as np
#saleprice correlation matrix
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(data[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
