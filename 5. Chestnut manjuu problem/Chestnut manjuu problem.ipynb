## [Problem] The day when chestnut buns cover the solar system
If you sprinkle a single chestnut bun with vine, how long will it take to cover the solar system with chestnut buns? Create a program to calculate this.

Make assumptions about the volume of the chestnut buns and the solar system based on your research and calculations. You may simplify the problem by not considering the gap between the chestnut buns and gravity.

Create a generic program that can be applied to other situations, such as the time it takes for the Tokyo Dome to fill up when you sprinkle a ball with Byvine. Also, please create a notebook that is easy to understand, e.g., display a graph.

The diameter of the solar system is approximately [287.46 billion kilometers](https://www.universetoday.com/15585/diameter-of-the-solar-system/#:~:text=Solar%20System%20a%20diameter%20of%20287.46%20billion%20km). meaning the radious is approximately:

Based on this [photo](https://u-taste.com/wp-content/uploads/2024/07/chestnut-buns.00_00_02_03.Still003_%E7%BB%93%E6%9E%9C-576x1024.webp) a Chestnut homey bun volume is approximetely 11x3x4=130 cm³

And the [Tokyo Dome Volume equals about 1.24 million m³](https://www.tokyo-dome.co.jp/en/dome/about/)

```python
import math
import matplotlib.pyplot as plt

volume_chestnut_bun_m3 = 1.3  
radius_solar_system = 14373000000000  
volume_solar_system = (4 / 3) * math.pi * radius_solar_system**3  
volume_tokyo_dome = 1240000  


def calculate_time_it_takes_fill_Byvine(initial_volume, target_volume, double_time=10):
    time = 0 #initial time
    times_array = [0] #initial array
    current_volume = initial_volume
    volumes_array = [current_volume]
    
    
    while current_volume < target_volume:
        time = time+double_time #every 10' seconds default will increase time
        current_volume *= 2
        volumes_array.append(current_volume)
        times_array.append(time)
    
    return time, volumes_array, times_array

time_solar_system, volumes_solar_system, times_solar_system = calculate_time_it_takes_fill_Byvine(volume_chestnut_bun_m3, volume_solar_system)
print(f"To fill the Solar System it will take: {time_solar_system} seconds.")
time_tokyo_dome, volumes_tokyo_dome, times_tokyo_dome = calculate_time_it_takes_fill_Byvine(volume_chestnut_bun_m3, volume_tokyo_dome)
print(f"To fill the Tokyo Dome it will take: {time_tokyo_dome} seconds.")


#GRAPHS
plt.figure(figsize=(12, 6))

#GRAPH SECONDS TO SOLAR SYSTEM
plt.subplot(1, 2, 1)
plt.plot(times_solar_system, volumes_solar_system, label="Solar System", color="blue")
plt.xlabel("Time (seconds)")
plt.ylabel("Volume (meters cubic)")
plt.title("Growth of Chestnut Bun Volume (Solar System)")
plt.legend()

#GRAPH SECONDS TO TOKYO DOME
plt.subplot(1, 2, 2)
plt.plot(times_tokyo_dome, volumes_tokyo_dome, label="Tokyo Dome", color="red")
plt.xlabel("Time (seconds)")
plt.ylabel("Volume (meters cubit)")
plt.title("Growth of Chestnut Bun Volume (Tokyo Dome)")
plt.legend()

plt.show()
```
