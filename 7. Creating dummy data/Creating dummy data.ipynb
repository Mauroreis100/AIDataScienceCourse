# Creating dummy data
When experimenting with programs related to machine learning, there are times when you want to have random data (dummy data) that is not real data. Let's get familiar with NumPy by creating dummy data with random numbers that follow a normal distribution.

## [Problem 1] Creating random numbers
Create 500 random numbers with a two-dimensional normal distribution whose mean is (-3, 0) and whose covariance matrix is
is


$$
\begin{pmatrix}
  1.0 & 0.8  \\
  0.8 & 1.0  \\ 
\end{pmatrix}
$$
Multidimensional normally distributed random numbers can be created by using np.random.multivariate_normal() .

The shape of the created ndarray will be (500, 2).

<Seed value>

To ensure that the value does not change from one run to the next, we will fix the seed value of the random number using np.random.seed() . We will use a seed value of 0.

np.random.seed(0)

```python
import numpy as np
np.random.seed(0)
mean = (-3, 0)
cov = [[1, 0.8], [0.8, 1]]
x = np.random.multivariate_normal(mean, cov, size=500)
x.shape

print(x)
```

## [Problem 2] Visualization by scatter plot
Visualize the data points created in Problem 1 with a scatter plot.

Scatter plots can be drawn by using matplotlib's plt.scatter() .

```python
import matplotlib.pyplot as plt
plt.grid()
plt.scatter(x[:, 0], x[:, 1])
```

## [Problem 3] Visualization by histogram
Visualize the data created in Problem 1 with a histogram.

A histogram can be drawn by using plt.hist().

matplotlib.pyplot.hist  Matplotlib 3.1.1 documentation

Create it for each dimension and create two histograms.
The range of the horizontal axis of each histogram should be aligned using plt.xlim() .

matplotlib.pyplot.xlim  Matplotlib 3.1.1 documentation

```python


# the histogram of the data
plt.hist(x[:, 0])
plt.hist(x[:, 1])
plt.title('Histogram of IQ')
plt.xlim(-10, 10)
plt.grid(True)
plt.show()
```

## [Problem 4] Addition of data
Create 500 new random numbers with a two-dimensional normal distribution whose mean is (0, -3) and whose covariance matrix is represented by the following.

$$
\begin{pmatrix}
  1.0 & 0.8  \\
  0.8 & 1.0  \\ 
\end{pmatrix}
$$
Then, visualize the data of problems 1 and 4 as one scatter plot. As a legend, show 0for Problem 1 and1for Problem 4.

<< Creation example >>

It is the same as the figure introduced at the beginning of assignment.

```python

meanAdd = [0, -3]
covAdd = [[1, 0.8],[0.8, 1]]
dataAdd = np.random.multivariate_normal(meanAdd, covAdd, size=500)
plt.title('Problem 1 vs Problem 4')
plt.scatter(x[:,0], x[:,1])
plt.scatter(dataAdd[:,0], dataAdd[:,1])
plt.axis('equal')
plt.legend(["Problem 1", "Problem 2"])
plt.show()
```

## [Problem 5] Data combination
It is better to keep the data together so that it is easier to deal with later. Combine the ndarrays created in Problems 1 and 4 into  a single ndarray of (1000, 2). The concatenation can be done using np.concatenate() or  np.vstack() .

```python
combinated_ndarrays=np.vstack((x,dataAdd))
print(combinated_ndarrays.shape)   
```

Create a ndarray of (1000, 3) by adding a new column of labels containing 0s and 1s.

```python
label_zero = np.zeros((500,1))
label_ones = np.ones((500,1))
new_zeros_ones=np.vstack((label_zero,label_ones))
to_new_array=np.hstack((combinated_ndarrays,new_zeros_ones))
print(to_new_array)
print(to_new_array.shape)
```
