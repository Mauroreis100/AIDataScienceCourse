# [Problem 1] Linear function
Consider the following formula where $x$ is in the range -50 to 50.

y=(1/2)x+1

Create a ndarray representing x at 0.1 intervals from -50 to 50, and a ndarray representing y for each element.

```python
import numpy as np

x = np.arange(-50, 50.1, 0.1) #Creating array with with spacing between values 0.1.
# or np.linespace(-50,59,1001)
y = (1/2)*x + 1
print(x.shape)
print(y.shape)
```

# [Problem 3] Find the gradient
Find the gradient at each x-coordinate. Here, the gradient is calculated by dividing the change in y by the change in x.



Each amount of change is approximately calculated from the difference between adjacent elements as shown in the figure below.


"Note"

* To obtain the gradient at each point exactly, we need to use the limit (derivative)
* The gradient uses the interval of the x-coordinates, so it is one less than the number of elements in x

```python
d_x=x[1:]-x[:-1]
d_y=y[1:]-y[:-1]
gradient=d_y/d_x
gradient
```

# [Problem 4] Draw a graph
Draw a graph of the linear function using Matplotlib. Also, draw a graph with x on the horizontal axis and gradient on the vertical axis.

```python
import matplotlib.pyplot as plt
plt.title('Gradient and Y')
plt.xlabel('X')
plt.xlabel('X')
plt.plot(x,y,label='y',ls='--')
plt.plot(x[1:],gradient, label='gradient',ls='--')
plt.legend()
plt.show()
```

# [Problem 5] Python functionalization
Please refer to the template to create a Python function that summarizes the problems 1 to 3.

Then, calculate the gradient of the following three equations at 0.1 intervals and output it with ndarray.

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x=np.arange(x_range[0],x_range[1],x_range[2])
    y=function(x)
    x_temp=x.reshape(x.shape[0],1)
    y_temp=y.reshape(y.shape[0],1)
    array_xy=np.hstack((x_temp,y_temp))
    d_x=x[1:]-x[:-1]
    d_y=y[1:]-y[:-1]
    gradient=d_y/d_x
    
    return array_xy, gradient
def function1(array_x):
    return array_x**2
def function2(array_x):
    return (2*(array_x**2)+2**array_x)
def function3(array_x):
    return np.sin(np.sqrt(array_x))

array_square, gradient_square = compute_gradient(function1)
plt.title('y=x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_square[:,0],array_square[:,1],label='y',ls='--')
plt.plot(array_square[:,0][1:],gradient_square,label='gradient',ls='--')
plt.legend()
plt.show
```

```python
array_exponential,gradient_exponential=compute_gradient(function2)
plt.title('y=2*x+2^x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_exponential[:,0],array_exponential[:,1],label='y',ls='--')
plt.plot(array_exponential[:,0][1:],gradient_exponential,label='gradient',ls='--')
plt.legend()
plt.show
```

```python
array_sin,gradient_sin=compute_gradient(function3,x_range=(0,50.1,0.1))
plt.title('y=sin(x/)')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_sin[:,0],array_sin[:,1],label='y',ls='--')
plt.plot(array_sin[:,0][1:],gradient_sin,label='gradient',ls='--')
plt.legend()
plt.show
```

```python
def compute_slope(arr,minIdx):
    before_min=arr[minIdx-1]
    at_min=arr[minIdx]
    after_min=arr[minIdx+1]
    slope_before=(at_min[1]-before_min[1])/(at_min[0]-before_min[0])
    slope_after=(after_min[1]-at_min[1])/(after_min[0]-at_min[0])
    return slope_before,slope_after
```
