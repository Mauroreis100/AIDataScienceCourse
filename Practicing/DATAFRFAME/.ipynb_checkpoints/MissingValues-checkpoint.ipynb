```python
import numpy as np
import pandas as pd
df = pd.read_csv("sales_data_contains_nan.csv")
df
```

# Deleting rows that have no value

```python
dropna_df = df.dropna()
dropna_df
```

# Complement missing values with specified values

```python
fillna_df = df.fillna(0) #Vai preencher todos os NaN com Zeros
fillna_df
```

You can also fill the missing value with the previous value by specifying the parameter method = 'ffill' in the fillna method.

```python
ffill_df = df.ffill()
ffill_df
```

# Using the whole table mean to fill empty blocks
Because of the date datatype we can't use the mean method to fill every NaN block

```python
df['newC']=''
df
```

```python
df['newC'] = pd.to_datetime(df['newC']) #Converting new added column to datetime
df
```

```python
df['日付'] = pd.to_datetime(df['日付']) #Changing the date column type to datetime
df
```

```python
df['newC']=df['日付'].astype(int)/1e9 #Converting the date column and inserting on the newC column
df
```

```python
df['newC'] = df['newC'].astype('float64')#Converting the newC column to a datatype that can do operations with other columns
df
```

```python
df_temporary=pd.DataFrame(df['日付']) #A temporary column to hold the YYYY-MM-DD format so we don't lose it
df_temporary
```

```python
df.drop('日付',axis=1,inplace=True) #For some reason the column did not drop without <inplace=True>
#df = df.drop(0, axis=1) Would have been fine too. I think...
df
```

```python
df = df.fillna(df.mean()) #Filling NaN with mean
df
```

```python
df.dtypes #Checking datatypes
```

```python
df.drop('newC',axis=1,inplace=True) #Deleting newC so the original date can come back
df
```

```python
df['日付']=df_temporary['日付'] #Replacing
df
```
