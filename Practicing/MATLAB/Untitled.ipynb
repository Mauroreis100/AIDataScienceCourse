```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x1=[1,2,3]
y1=[2,3,4]
x2=[7,6,5]
y2=[4,3,1]

fig, ax = plt.subplots(2, 2)
ax[0,0].plot(x1, y1)
ax[0,1].plot(x1, y2)
ax[1,0].plot(x2, y1)
ax[1,1].plot(x2, y2)
plt.show()
```

```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

fig, axes = plt.subplots(1, 2)
axes[0].plot(x1, y1)
axes[1].plot(x2, y2)
axes[0].set_title('Subplot1 title')
axes[1].set_title('Subplot2 title')
fig.suptitle('Graph title')
plt.show()
```
